```python
import numpy as np
from skimage.draw import line as draw_line
import matplotlib.pyplot as plt
```

```python
# Load data from text file
with open('input/day5.txt') as f:
    data_string = f.read()
split_data = data_string.split('\n')

split_data = [line.replace(' -> ', ',') for line in  split_data]
split_data = [line.split(',') for line in split_data]
split_data = np.array(split_data).astype(int)

print(split_data)
```

## Part 1

```python
img = np.zeros((1000, 1000))
                   
for line in split_data:
    # If horizontal or vertical
    if (line[0]==line[2]) or (line[1]==line[3]):
        rr, cc = draw_line(line[1], line[0], line[3], line[2])
        img[rr, cc] += 1
```

```python
plt.figure(figsize=(12,12))
plt.imshow(img)
```

```python
len(img[img>=2])
```
